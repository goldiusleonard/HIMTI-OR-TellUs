```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import os
import re
from sklearn.preprocessing import LabelEncoder
```

```python
news = []
print("News Directory:")
for dirname, _, filenames in os.walk('./data/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        df = pd.read_csv(os.path.join(dirname, filename), index_col = None, header = 0)
        row, cols = df.shape
        news.append(df)
        
data_set = pd.concat(news, axis=0, ignore_index=True)
del news

data_set = data_set.drop('Unnamed: 0', axis = 1)
print(data_set)
```

```python
labelEncoder_X = LabelEncoder()
data_set['news_category'] = labelEncoder_X.fit_transform(data_set['news_category'])
data_set.head()
```

```python
corpus = []
ps = PorterStemmer()
for i in range(0, len(data_set)):
    review = data_set["news_article"][i]
    review = re.sub("[^a-zA-Z]", " ", review)
    review = review.lower()
    review = review.split()
    review = [word for word in review if not word in set(stopwords.words("english"))]
    review = [ps.stem(word1) for word1 in review]
    review = " ".join(review)
    corpus.append(review)

print(corpus)
```

```python
txt = open("corpus.txt", "w")
for index in corpus:
    txt.write(index)
    txt.write("\n")
txt.close()
```

```python
txt = open("corpus.txt", "r")
corpus = []
for line in txt:
    sentence = line.rstrip("\n")
    corpus.append(sentence)

print(corpus)
```
